```python
clas = MLPClassifier(solver='adam', alpha=0.0004, hidden_layer_sizes=(50, 30, 10, 5), random_state=0)
clas.fit(X, Y['0'].tolist())
```

```python
clas.predict_proba(Z[:])
```

```python
pred=clas.predict(Z)
df_pred=pd.DataFrame(data=pred)
```
