```python
import numpy as np
import pandas as pd
import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import scipy
from skimage import feature
from scipy.stats import skew
import skimage
```

```python
train=r"C:\Users\DELL\Downloads\Ravelling_dataset\mod_ravelling_dataset\train"
test=r"C:\Users\DELL\Downloads\Ravelling_dataset\mod_ravelling_dataset\test"
```

```python
def dataset_creation(train):
    img=[]
    for file in os.listdir(train):
            
            image_path = os.path.join(train, file)
            image = np.array(Image.open(image_path))
            image = np.resize(image, (80, 80, 3))
            image = image.astype('float32')
            
            img.append(image)
    return img
```

```python
test_id = dataset_creation(test)
```

```python
len(test_id)
```

```python
#Mean
red_mean_arr=[]
green_mean_arr=[]
blue_mean_arr=[]
for i in range(300):
    # Feature 1
    red_mean_arr.append(train_id[i][:,:,0].mean())
    # Feature 2
    green_mean_arr.append(train_id[i][:,:,1].mean())
    # Feature 3
    blue_mean_arr.append(train_id[i][:,:,2].mean())
```

```python
#STD Dev
red_sd_arr=[]
green_sd_arr=[]
blue_sd_arr=[]
for i in range(300):
    # Feature 4
    red_sd_arr.append(np.std(train_id[i][:,:,0]))
    # Feature 5
    green_sd_arr.append(np.std(train_id[i][:,:,1]))
    # Feature 6
    blue_sd_arr.append(np.std(train_id[i][:,:,2]))
```

```python
#Skewness
red_sk_arr=[]
green_sk_arr=[]
blue_sk_arr=[]
for i in range(300):
    # Feature 7
    red_sk_arr.append(scipy.stats.skew(train_id[i][:,:,0].reshape(6400,1), axis=0))
    # Feature 8
    green_sk_arr.append(scipy.stats.skew(train_id[i][:,:,1].reshape(6400,1), axis=0))
    # Feature 9
    blue_sk_arr.append(scipy.stats.skew(train_id[i][:,:,2].reshape(6400,1), axis=0))
```

```python
#Kurtosis
red_ku_arr=[]
green_ku_arr=[]
blue_ku_arr=[]
for i in range(300):
    # Feature 7
    red_ku_arr.append(scipy.stats.kurtosis(train_id[i][:,:,0].reshape(6400,1), axis=0))
    # Feature 8
    green_ku_arr.append(scipy.stats.kurtosis(train_id[i][:,:,1].reshape(6400,1), axis=0))
    # Feature 9
    blue_ku_arr.append(scipy.stats.kurtosis(train_id[i][:,:,2].reshape(6400,1), axis=0))
```

```python
#Entropy
red_en_arr=[]
green_en_arr=[]
blue_en_arr=[]
for i in range(300):
    # Feature 13
    red_en_arr.append(scipy.stats.entropy(train_id[i][:,:,0].reshape(6400,1), axis=0))
    # Feature 14
    green_en_arr.append(scipy.stats.entropy(train_id[i][:,:,1].reshape(6400,1), axis=0))
    # Feature 15
    blue_en_arr.append(scipy.stats.entropy(train_id[i][:,:,2].reshape(6400,1), axis=0))
```

```python
#Range
red_r_arr=[]
green_r_arr=[]
blue_r_arr=[]
for i in range(300):
    # Feature 16
    red_r_arr.append(np.ptp(train_id[i][:,:,0].reshape(6400,1)))
    # Feature 17
    green_r_arr.append(np.ptp(train_id[i][:,:,1].reshape(6400,1)))
    # Feature 18
    blue_r_arr.append(np.ptp(train_id[i][:,:,2].reshape(6400,1)))
```

```python
#GLCM
```

```python
g0=[]
glcm_asm_0=[]
glcm_contrast_0=[]
glcm_correlation_0=[]
glcm_diss_0=[]

for i in range(300):
    g0.append(skimage.feature.texture.greycomatrix(skimage.color.rgb2gray(train_id[i]).astype('uint8'), [1], [0], levels=256))
    # Feature 19
    glcm_asm_0.append(skimage.feature.texture.greycoprops(g0[i], prop='ASM'))
    # Feature 20
    glcm_contrast_0.append(skimage.feature.texture.greycoprops(g0[i], prop='contrast'))
    # Feature 21
    glcm_correlation_0.append(skimage.feature.texture.greycoprops(g0[i], prop='correlation'))
    # Feature 22
    glcm_diss_0.append(skimage.feature.texture.greycoprops(g0[i], prop='dissimilarity'))
```

```python
gpi4=[]
glcm_asm_pi4=[]
glcm_contrast_pi4=[]
glcm_correlation_pi4=[]
glcm_diss_pi4=[]

for i in range(300):
    gpi4.append(skimage.feature.texture.greycomatrix(skimage.color.rgb2gray(train_id[i]).astype('uint8'), [1], [np.pi/4], levels=256))
    # Feature 19
    glcm_asm_pi4.append(skimage.feature.texture.greycoprops(gpi4[i], prop='ASM'))
    # Feature 20
    glcm_contrast_pi4.append(skimage.feature.texture.greycoprops(gpi4[i], prop='contrast'))
    # Feature 21
    glcm_correlation_pi4.append(skimage.feature.texture.greycoprops(gpi4[i], prop='correlation'))
    # Feature 22
    glcm_diss_pi4.append(skimage.feature.texture.greycoprops(gpi4[i], prop='dissimilarity'))
```
